# perceptron

# importing dataset from library


```python
from sklearn.datasets import load_iris
```

#importing perceptron model

```python
from sklearn.linear_model import Perceptron
```

importing a performance metric


```python
from sklearn.metrics import accuracy_score
```

#loading dataset

```python
iris =load_iris()
```

#printing dataset

```python
print(iris)
```

#independent and dependent variables

```python
x=iris.data[:,(2,3)]
y=(iris.target ==0).astype(int)
```

0-sepal length,1-sepal width,2-petallength,3-petal width last column-species(setosa,versicolour&viriginica)

```python
x
```

```python
y
```

# applying the model

```python
ptron = Perceptron(random_state=42)
```

#fitting values to model

```python
ptron.fit(x,y)
```

#predicting using model

```python
y_pred=ptron.predict(x)
```

```python
y_pred
```

#accuracy of model

```python
print(f'Accuracy Score: {accuracy_score(y,y_pred)}')
```

# cefficient value from model

```python
print(f'Coefficient Values: {ptron.coef_}')
```

# intercept value

```python
print(f'Intercept Values: {ptron.intercept_}')
```

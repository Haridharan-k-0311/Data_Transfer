```python
import pandas as pd 
import numpy as nm
import matplotlib.pyplot as mtp
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('user_data.csv')
```

```python
x=data.iloc[:,[2,3]]
y = data.iloc[:, 4].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
impt=StandardScaler()
x_train=impt.fit_transform(x_train)
x_test=impt.fit_transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
scl=DecisionTreeClassifier(criterion='entropy',random_state=0)
```

```python
scl.fit(x_train,y_train)
```

```python
y_pred=scl.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix  
cm= confusion_matrix(y_test, y_pred) 
```

```python
from matplotlib.colors import ListedColormap 
```

```python
x_set, y_set = x_train, y_train  

x1, x2 = nm.meshgrid(nm.arange(start=x_set[:, 0].min() - 1, stop=x_set[:, 0].max() + 1, step=0.01),  
                      nm.arange(start=x_set[:, 1].min() - 1, stop=x_set[:, 1].max() + 1, step=0.01)) 

mtp.contourf(x1, x2, scl.predict(nm.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),  
             alpha=0.75, cmap=ListedColormap(('purple', 'green')))  

mtp.xlim(x1.min(), x1.max())  
mtp.ylim(x2.min(), x2.max())  

for i, j in enumerate(nm.unique(y_set)):  
    mtp.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],  
                c=ListedColormap(('purple', 'green'))(i), label=j)  

mtp.title('Decision Tree Algorithm (Training set)')  
mtp.xlabel('Age')  
mtp.ylabel('Estimated Salary')  
mtp.legend()  
mtp.show()
```

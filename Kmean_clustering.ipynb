```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
```

```python
# Generate synthetic data
X, _ = make_blobs(n_samples=400, centers=4, random_state=42)
```

```python
# Split the data into training and test sets
X_train, X_test = train_test_split(X, test_size=0.25, random_state=42)
```

```python
# Create KMeans model with 4 clusters
kmeans = KMeans(n_clusters=4, random_state=42)
```

```python
# Fit the model on training data
kmeans.fit(X_train)
```

```python
# Predict the clusters for test data
test_clusters = kmeans.predict(X_test)
```

```python
# Plot the training data and the centroids
plt.figure(figsize=(8, 6))
plt.scatter(X_train[:, 0], X_train[:, 1], c=kmeans.predict(X_train), cmap='viridis', marker='o', edgecolor='k', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', s=200)
plt.title("K-Means Clustering - Training Data")
plt.show()
```

```python
# Plot the test data with predicted clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_test[:, 0], X_test[:, 1], c=test_clusters, cmap='viridis', marker='o', edgecolor='k', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', s=200)
plt.title("K-Means Clustering - Test Data")
plt.show()
```

```python
# Print K-Means results as text output
print("K-Means Clustering Results")
print("----------------------------")
print(f"Number of clusters: {kmeans.n_clusters}")
print(f"Centroids of the clusters (4 clusters):\n{kmeans.cluster_centers_}")
print("\nCluster assignments for the training data:")
print(kmeans.predict(X_train))
print("\nCluster assignments for the test data:")
print(test_clusters)
```

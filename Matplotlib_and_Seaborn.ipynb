```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# Load a sample dataset from Seaborn
tips = sns.load_dataset('tips')
```

# --- Matplotlib Examples ---

```python
# 1. Line Plot (Simple Line Plot)
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Line')
plt.show()
```

```python
# 2. Scatter Plot (Simple Scatter Plot)
x = np.random.rand(50)
y = np.random.rand(50)
plt.figure(figsize=(8, 6))
plt.scatter(x, y)
plt.show()
```

```python
# 3. Bar Plot (Bar Chart)
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]
plt.figure(figsize=(8, 6))
plt.bar(categories, values)
plt.show()
```

```python
# 4. Histogram (Histogram Plot)
data = np.random.randn(1000)
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, edgecolor='black')
plt.show()
```

```python
# 5. Pie Chart (Pie Plot)
sizes = [30, 20, 25, 25]
labels = ['A', 'B', 'C', 'D']
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
```

# --- Seaborn Examples ---


```python
# 1. Scatter Plot with Seaborn (with Hue)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=tips)
plt.show()
```

```python
# 2. Box Plot (Boxplot)
plt.figure(figsize=(8, 6))
sns.boxplot(data=tips)
plt.show()
```

```python
# 3. Violin Plot (Violin Plot)
plt.figure(figsize=(8, 6))
sns.violinplot(data=tips)
plt.show()
```

```python
# 4. Heatmap (Correlation Heatmap)
# Select only numerical columns for the correlation matrix
numerical_tips = tips.select_dtypes(include=[np.number])

# Create a correlation matrix from the numerical columns
corr_matrix = numerical_tips.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```

```python
# 5. Pair Plot (Pairwise Plot)
plt.figure(figsize=(8, 6))
sns.pairplot(tips, hue='sex', palette='Set1')
plt.show()
```

```python
# 6. Bar Plot (Bar Chart using Seaborn)
plt.figure(figsize=(8, 6))
sns.barplot(x='day', y='total_bill', data=tips, palette='Blues')
plt.show()
```

```python
# 7. Histogram (Histogram Plot using Seaborn)
plt.figure(figsize=(8, 6))
sns.histplot(data, bins=30, kde=True, color='orange')
plt.show()
```
